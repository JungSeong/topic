# 클래스, 상속, 오버라이딩, super()

클래스 : 객체를 찍어내주는 틀

상속 : 부모 클래스를 그대로 물려 받기

오버라이딩 : 덮어쓰기

super() : 오버라이딩이 되는 상황에서 상속받을 속성들은 전부 받고 싶을 때 사용

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def draw_sin(A, t, f, b):
    y = A * np.sin(2*np.pi*f*t) + b

    plt.figure(figsize=(12,6))
    plt.plot(t, y)
    plt.grid()
    plt.show()
```

```python
t = np.arange(0, 3, 0.01)
draw_sin(2, t, 2, 2)
```

```python
class DrawSin:
    def __init__(self, amp, time, freq, bias):
        self.amp = amp
        self.time = np.arange(0, time, 0.01)
        self.freq = freq
        self.bias = bias

    def calc_sin(self):
        return self.amp * np.sin(2*np.pi*self.freq*self.time) + self.bias

    def draw_sin(self):
        plt.figure(figsize=(12,6))
        plt.plot(self.time, self.calc_sin())
        plt.grid()
        plt.show()
```

```python
sin1 = DrawSin(3, 3, 3, 0)
sin1.draw_sin()
```

```python
sin1.bias = 5
sin1.draw_sin()
```

```python
sin2 = DrawSin(5, 5, 1, 1)
sin2.draw_sin()
```

```python
# DrawSin의 모든 클래스를 상속(Inheritance) 받음

class DrawSinusoidal(DrawSin):
    def calc_cos(self):
        return self.amp * np.cos(2*np.pi*self.freq*self.time) + self.bias

    def draw_cos(self):
        plt.figure(figsize=(12,6))
        plt.plot(self.time, self.calc_cos())
        plt.grid()
        plt.show()
```

```python
cos1 = DrawSinusoidal(3, 3, 3, 0)
cos1.draw_cos()
```

```python
cos1.__dir__()
```

```python
sin3 = DrawSinusoidal(3, 3, 3, 0)
sin3.draw_sin()
```

```python
# Override

class DrawSinusoidal2(DrawSinusoidal):
    def draw_sin(self):
        plt.figure(figsize=(12,6))
        plt.plot(self.time, self.calc_sin())
        plt.grid()
        plt.xlabel('time (sec)')
        plt.ylabel('Sin')
        plt.title('Sin Graph')
        plt.show()

    def draw_cos(self):
        plt.figure(figsize=(12,6))
        plt.plot(self.time, self.calc_cos())
        plt.grid()
        plt.xlabel('time (sec)')
        plt.ylabel('Cos')
        plt.title('Cosine Graph')
        plt.show()
```

```python
sin4 = DrawSinusoidal2(3, 3, 3, 0)
sin4.draw_sin()
sin4.draw_cos()
```

```python
# DrawSinusoidal2의 __init__이 오버라이딩된 것이기 때문에 __init__의 내용이 없어진다.

class DrawSinusoidal3(DrawSinusoidal2):
    def __init__(self, amp, time, freq, bias, ts):
        self.ts = ts
```

```python
ds3 = DrawSinusoidal3(1, 1, 1, 3, 0.01)
ds3.__dict__
```

```python
# super() : 오버라이딩이 되는 상황에서 상속받을 속성들은 전부 받고 싶을 때 사용

class DrawSinusoidal3(DrawSinusoidal2):
    def __init__(self, amp, time, freq, bias, ts):
        super().__init__(amp, time, freq, bias)
        self.ts = ts
        self.time = np.arange(0, time, self.ts)
```

```python
ds3 = DrawSinusoidal3(3, 1, 1, 3, 0.1)
ds3.__dict__
```

```python
ds3.draw_sin()
ds3.draw_cos()
```
